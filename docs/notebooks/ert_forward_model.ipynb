Build a resistivity model with a buried block, simulate a dipole-dipole survey,
and view the apparent-resistivity pseudosection with the open-source
[pyGIMLi](https://www.pygimli.org/) library.


## Install pyGIMLi (Colab)

The first run takes a few minutes. On Colab, install with pip:


```python
import importlib.util
import subprocess
import sys

if importlib.util.find_spec('pygimli') is None:
    if 'google.colab' not in sys.modules:
        raise RuntimeError('pygimli is required. Run this notebook in Colab or install it with: pip install pygimli')
    subprocess.check_call([sys.executable, '-m', 'pip', 'install', '-q', 'pygimli'])

import pygimli as pg
import pygimli.physics.ert as ert
import numpy as np
```

## Survey geometry and mesh

A dipole-dipole array of 24 electrodes at 2 m spacing.


```python
scheme = ert.createData(elecs=np.linspace(0, 46, 24), schemeName='dd')
world = pg.meshtools.createWorld(start=[-10, -20], end=[56, 0], marker=1)
block = pg.meshtools.createRectangle(start=[18, -8], end=[30, -3], marker=2)
geom = world + block
mesh = pg.meshtools.createMesh(geom, quality=33, area=1.0)
```

## Assign resistivities and forward model

Background 100 ohm-m, conductive block 10 ohm-m.


```python
rhomap = [[1, 100.0], [2, 10.0]]
data = ert.simulate(mesh, res=rhomap, scheme=scheme, noiseLevel=0.02, seed=1)
ert.show(data)   # apparent-resistivity pseudosection
```
