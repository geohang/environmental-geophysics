Compute the gravity anomaly of a buried sphere and explore how depth, radius,
and density contrast change the signal. This is the code version of the
*Buried-Body Anomaly Modeler* demo.


```python
import numpy as np
import matplotlib.pyplot as plt
G = 6.674e-11  # m^3 kg^-1 s^-2
```

## Sphere anomaly

$g_z(x) = \dfrac{G\,\frac{4}{3}\pi R^3 \Delta\rho\; z}{(x^2+z^2)^{3/2}}$, converted to mGal ($\times 10^5$).


```python
def sphere_gz(x, z, R, drho):
    M = (4/3)*np.pi*R**3*drho          # kg
    return 1e5 * G * M * z / (x**2 + z**2)**1.5   # mGal

x = np.linspace(-100, 100, 401)
g = sphere_gz(x, z=30, R=10, drho=400)
print(f'peak anomaly = {g.max():.4f} mGal')
```

```python
plt.figure(figsize=(7,4))
plt.plot(x, g)
plt.xlabel('distance (m)'); plt.ylabel('g_z (mGal)')
plt.title('Gravity anomaly of a buried sphere'); plt.grid(alpha=0.3); plt.show()
```

## Half-width depth rule

For a sphere the half-width relates to depth by $z \approx x_{1/2}/0.766$. Verify it below.


```python
half = g.max()/2
x_half = x[(x>0) & (g<=half)][0]
print(f'x_1/2 = {x_half:.1f} m  ->  depth = {x_half/0.766:.1f} m (true = 30 m)')
```
