A relative gravimeter reading drifts over time because the internal spring relaxes.
We remove that drift by revisiting a base station and assuming the drift is linear
between base readings. This notebook reproduces the drift-correction activity in code.


```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Load the survey loop

Each row is one occupation: time (minutes from start), station id, and raw reading (mGal).
Station `BASE` is reoccupied at the start, middle, and end.


```python
# time_min, station, reading_mGal
t     = np.array([0,  12,  25,  38,  52,  65,  80])
stn   = np.array(['BASE','A','B','BASE','C','D','BASE'])
g_raw = np.array([1000.00, 1002.31, 1004.85, 1000.07, 1003.60, 1005.9, 1000.14])
```

## 2. Fit the drift from base-station reoccupations

The base readings should be identical; any change is drift. Fit a straight line
through the base occupations and subtract it from every reading.


```python
base = stn == 'BASE'
coef = np.polyfit(t[base], g_raw[base], 1)   # slope (mGal/min), intercept
drift = np.polyval(coef, t)
g_corr = g_raw - (drift - drift[0])
print(f'drift rate = {coef[0]*60:.3f} mGal/hour')
```

## 3. Compare raw vs. corrected


```python
fig, ax = plt.subplots(figsize=(7,4))
ax.plot(t, g_raw, 'o--', label='raw')
ax.plot(t, g_corr, 's-', label='drift-corrected')
ax.set_xlabel('time (min)'); ax.set_ylabel('gravity (mGal)')
ax.legend(); ax.grid(alpha=0.3); plt.show()
```
